```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
img = imread('segmentation_image.jpg')
```

```python
plt.imshow(img)
```

```python
img = img/255 #scaling
img.shape
```

```python
X = img.reshape(-1,3)
```

```python
seg = KMeans(n_clusters=3)
seg.fit(X)
```

```python
n_img = seg.cluster_centers_
n_img = n_img[seg.labels_]
```

```python
n_img = n_img.reshape(311, 553, 3)
plt.imshow(n_img)
```
